```python
import pandas as pd
import numpy as  np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')

```

```python
df=pd.read_csv('./titanic.csv')
df.info()
```

```python
plt.figure(figsize=(8,6))

sns.countplot(data=df,x='Survived',hue='Pclass',color=('red'))
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.countplot(data=df,x='Embarked',hue='Sex',color='yellow')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(data=df,x='Pclass',hue='Embarked',color='red')
plt.title('CLASS---Embarked---GRAPH')
plt.show()
```

```python
df.info()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Pclass',hue='SibSp',color='green')
plt.title('CLASS---SIBSP---GRAPH')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Pclass',hue='Parch',color='green')
plt.title('CLASS---PARCH---GRAPH')
plt.show()
```

```python
#----class---- age----sex---

plt.figure(figsize=(6,4))
sns.barplot(data=df,x='Pclass',y='Age',hue='Sex',color='pink')
plt.title('---PCLASS--Sex--Age---GRAPH')
plt.show()
```

```python
#-------class---age-----location----

plt.figure(figsize=(6,4))
sns.barplot(data=df,x='Pclass',y='Age',color='red',hue='Embarked')
plt.title('---PCLASS--Age-Embarked--GRAPH')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df,x='Pclass',y='Fare',color='red',hue='Embarked')
plt.title('---PCLASS--Fare-Embarked--GRAPH')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df,x='Pclass',y='Fare',color='red',hue='Sex')
plt.title('---PCLASS--Sex--Fare---GRAPH')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df,x='Pclass',y='Age',hue='SibSp',color='red')
plt.title('---PCLASS--Age--SibSp---GRAPH')
plt.show()
```
