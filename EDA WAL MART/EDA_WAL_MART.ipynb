```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('./walmart-sales-dataset-of-45stores.csv',parse_dates=['Date'])
df.head()
```

```python
df.Weekday.replace({0:'SUN',1:'MON',2:'TUE',3:'WED',4:'THU',5:'FRI',6:'SAT'},inplace=True)
df
```

# Temperature coversion (-3F--32)X5/9

```python
df['celsius']=[(Temp-32)*(5/9) for Temp in df.Temperature]
```

# To create new column Cold and Warm based on celsius

```python
df['Heat'] =pd.cut(x=df.celsius, bins=[-20,10,35],labels=['Cold','Warm'])
```

```python
df.head()
```

```python
plt.figure(facecolor='gray',figsize=(10,8))
sns.barplot(data=df,x='Heat',y='Weekly_Sales',hue='Holiday_Flag')
plt.show()
```
